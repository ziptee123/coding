```python
cart=[]
while True:
  print("==選項==")
  print("1.新增物品")
  print("2.查看購物車")
  print("3.結帳")
  choice = input("請輸入選項 (1, 2, 3)：")
  if choice=="1":
    add=input("想買什麼?")
    cart.append(add)
    print(f"已經將{add}加入購物車。")
  elif choice=="2":
    print(f"目前共有{len(cart)}項物品!")
    print(f"你總共買了:{cart}")
  elif choice=="3":
    print("歡迎下次光臨!")
    break
  else:
    print("沒有這個選項!")

```

```python
餘額=1000
while True:
  print("==選項==")
  print("1.查詢餘額 2.存錢 3.提款 4.退出")
  choice=input("請輸入選項(1,2,3,4):")
  if choice=="1":
    print(f"目前餘額為{餘額}元!")
  elif choice=="2":
    add=int(input("請輸入存款金額:"))
    餘額=餘額+(add)
    print(f"已存入{add}元!")
    print(f"目前餘額為{餘額}元!")
  elif choice=="3":
    det=int(input("請輸入提款金額:"))
    if det>餘額:
      print("餘額不足!")
    else:
      餘額=餘額-(det)
      print(f"已提款{det}元!")
      print(f"目前餘額為{餘額}元!")
  elif choice=="4":
    print("請拿回信用卡。")
    break
  else:
    print("沒有這個選項!")
```

```python
slime_hp=50
while slime_hp>0:
  print("1.普通攻擊" "2.大招")
  attack=input("使用技能(1、2)")
  if attack=="1":
    slime_hp=slime_hp-5
    print(f"史萊姆受到傷害!目前血量:{slime_hp}")
  elif attack=="2":
    slime_hp=slime_hp-10
    print(f"史萊姆受到傷害!目前血量:{slime_hp}")
  else:
    slime_hp=slime_hp-0
    print(f"技能無效!目前血量:{slime_hp}")
print("戰鬥結束!")
```

```python
menu={"紅茶":30,"奶茶":45,"綠茶":25}
total=0
while True:
  print(menu)
  order=input("請問要喝什麼?(按下結帳以結束)")
  if order in menu:
    total=total+menu[order]
    print(f"總花費{total}元。")
  elif order=="結帳":
    break
  else:
    print("沒有賣這個喔!")
print(f"你總共要花{total}元。")
```
