```python
餘額=1000
while True:
  print("==選項==")
  print("1.查詢餘額 2.存錢 3.提款 4.退出")
  choice=input("請輸入選項(1,2,3,4):")
  if choice=="1":
    print(f"目前餘額為{餘額}元!")
  elif choice=="2":
    add=int(input("請輸入存款金額:"))
    餘額=餘額+(add)
    print(f"已存入{add}元!")
    print(f"目前餘額為{餘額}元!")
  elif choice=="3":
    det=int(input("請輸入提款金額:"))
    if det>餘額:
      print("餘額不足!")
    else:
      餘額=餘額-(det)
      print(f"已提款{det}元!")
      print(f"目前餘額為{餘額}元!")
  elif choice=="4":
    print("請拿回信用卡。")
    break
  else:
    print("沒有這個選項!")
```

```python
import random
secret=random.randint(1,100)
history=[]
while True:
  try:
    guess=int(input("請猜一個數字:"))
    history.append(guess)
    if guess>secret:
      print("太大了!")
    elif guess<secret:
      print("太小了!")
    else :
      print("正確!")
      break
  except ValueError:
    print("請輸入數字!")
print(f"你總共猜了這些數字:{history}")
print(f"你總共猜了{len(history)}次!")
```

```python
import random
import time
def dice():
  dice=random.randint(1,6)
  return dice
pocket=100
while True:
  try:
    money=int(input("請輸入你要下注的金額:(輸入0離開)"))
    if money==0:
      print("你不下注。選擇離開。")
      break
    player=dice()
    opponent=dice()
    print("擲骰子中...")
    time.sleep(1)
    print(f"你擲的骰子是{player},對方擲的骰子是{opponent}")
    if player>opponent:
      print("你贏了!")
      pocket=pocket+money
      print(f"你現在有{pocket}元")
    elif player<opponent:
      print("你輸了!")
      pocket=pocket-money
      print(f"你現在有{pocket}元")
    else:
      print("平手!")
      print(f"你現在有{pocket}元")
  except ValueError:
    print("請輸入數字!")
```

```python
import random
import time
def upg(level):
  cho=random.randint(1,100)
  if cho<=70:
    print("強化成功!等級加1。")
    level=level+1
  else:
    print("強化失敗。")
    if level>1:
      print("等級減1。")
      level=level-1
    else:
      print("等級不變。")
  return level
level=1
while True:
  try:
    tis=int(input("請問你要強化幾次?(輸入0離開)"))
    if tis==0:
      print("不升級，已離開。")
      break
    for i in range(tis):
      print(f"第{i+1}次升級中...")
      time.sleep(0.5)
      level=upg(level)
    print(f"你現在的等級是{level}")
  except ValueError:
    print("請輸入數字!")
```

```python
import random
def add():
  number1=random.randint(1,50)
  number2=random.randint(1,50)
  addi=number1+number2
  print(f"請計算這道題目:{number1}+{number2}")
  return addi
addi = add()
while True:
  try:
    ans=int(input("請輸入答案。"))
    if ans!=addi:
      print("回答錯誤!請重新作答。")
    else:
      print("回答正確，登入成功。")
      break
  except ValueError:
    print("請輸入數字!")
```

```python
import random
import time
ans = random.randint(1,10)
print("炸彈已啟動！5秒內輸入正確密碼解除。")
start = time.time()
while True:
  if time.time()-start > 5:
      print("時間到！炸彈爆炸！")
      break
  try:
    ito = int(input("請輸入密碼:"))
    if ito == ans:
      print("回答正確! 炸彈已解除。")
      break
    else:
      print("密碼錯誤!")
  except ValueError:
      print("輸入無效! 炸彈爆炸!")
      break
```
