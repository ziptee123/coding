```python
import random
secret=random.randint(1,100)
history=[]
while True:
  try:
    guess=int(input("請猜一個數字:"))
    history.append(guess)
    if guess>secret:
      print("太大了!")
    elif guess<secret:
      print("太小了!")
    else :
      print("正確!")
      break
  except ValueError:
    print("請輸入數字!")
print(f"你總共猜了這些數字:{history}")
print(f"你總共猜了{len(history)}次!")
```

```python
import random
import time
def dice():
  dice=random.randint(1,6)
  return dice
pocket=100
while True:
  try:
    money=int(input("請輸入你要下注的金額:(輸入0離開)"))
    if money==0:
      print("你不下注。選擇離開。")
      break
    player=dice()
    opponent=dice()
    print("擲骰子中...")
    time.sleep(1)
    print(f"你擲的骰子是{player},對方擲的骰子是{opponent}")
    if player>opponent:
      print("你贏了!")
      pocket=pocket+money
      print(f"你現在有{pocket}元")
    elif player<opponent:
      print("你輸了!")
      pocket=pocket-money
      print(f"你現在有{pocket}元")
    else:
      print("平手!")
      print(f"你現在有{pocket}元")
  except ValueError:
    print("請輸入數字!")
```
