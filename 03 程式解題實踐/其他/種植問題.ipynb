```python
#種植問題
#改編自2021 資訊學科能力競賽 北二區複賽
x1, y1, x2, y2, x3, y3, x4, y4 = map(int, input().split())
px, py = map(int, input().split())
bxmin = min(x1, x2)
bxmax = max(x1, x2)
bymin = min(y1, y2)
bymax = max(y1, y2)
cxmin = min(x3, x4)
cxmax = max(x3, x4)
cymin = min(y3, y4)
cymax = max(y3, y4)
#可以簡化成
#def getra(v1, v2):
  #return min(v1, v2), max(v1, v2)
#bxmin, bxmax = getra(x1, x2)
#bymin, bymax = getra(y1, y2)
#cxmin, cxmax = getra(x3, x4)
#cymin, cymax = getra(y3, y4)
#更精簡，也比較不容易錯
if cxmin < px < cxmax and cymin < py < cymax:
    print("3")
elif bxmin < px < bxmax and bymin < py < bymax:
    print("2")
else:
    print("1")
```
