````python
#a009. 解碼器
#先根據Sample Input跟Output算出它們的差值(7)
#相對應的密碼的ASCII值-7=相對應的解密後得出來的密碼的ASCII值
s=input().strip()
result=""
for i in s:
  result+=chr(ord(i)-7)
  #ord()將字元轉成數字，進行完位移運算後，再用chr()將數字轉回字元
print(result)
````

````python
#a013. 羅馬數字
def roam(s):
    tran = {"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000,"IV":4,"IX":9,"XL":40,"XC":90,"CD":400,"CM":900}
    i = 0
    total = 0
    while i < len(s):
        if i + 1 < len(s) and s[i:i+2] in tran:
            total += tran[s[i:i+2]]
            i += 2
        else:
            total += tran[s[i]]
            i += 1
    return total
def roamit(num):
    v = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    sy = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    res = ""
    j = 0
    while num > 0:
        for k in range(num // v[j]):
            res += sy[j]
            num -= v[j]
        j += 1
    return res
while True: #主程式碼
    ab = input().strip()
    if ab == "#":
        break
    a, b = ab.split()
    na = roam(a)
    nb = roam(b)
    n = abs(na - nb)
    if n == 0:
        print("ZERO")
    else:
        print(roamit(n))
````

特殊規則：IV、IX 這類特殊組合要額外寫出來。  
邏輯檢查：

s[i:i+2] 則是檢查是否兩個字元的組合有沒有出現在上面的字典裡。  
2. 阿拉伯數字轉羅馬數字 (roamit)  
  使用貪婪演算法:將數值由大到小排列，而且順序對應。  
  這種設計可以透過同一個索引值i同時存取阿拉伯數字與其對應的羅馬符號。  
```python
  while num>0:   
    for k in range(num//v[j]):
```  
  這裡的while可以確保程式會持續處理直到數值被扣完為止。

````python
#a065. 提款卡密碼
n=input()
s=""
for i in range(6): #密碼6位
  a=str(abs(ord(n[i])-ord(n[i+1])))
  #算出兩個字母所對應的ASCII碼差的絕對值即為第i位密碼
  #用str()轉換成字串，否則數字會加起來
  s+=a
print(s)
````

````python
#c009. 10473 - Simple Base Conversion
while True:
  n=input().strip()
  if n[0]=="-":
    break
  if n.startswith("0x"):
    print(int(n,16)) #16進位轉10進位
  else:
    print("0x"+format(int(n),"X")) #10進位轉16進位
````

````python
#c044. 10008 - What's Cryptanalysis
n=int(input())
text=""
for i in range(n):
  engli=input()
  for j in engli:
    if j.isalpha(): #判斷是否為英文字母
      text=text+j.upper() #全部變大寫
eng={} #使用字典，程式可以只針對實際出現過的字母建立空間
for e in text:
  if e in eng:
    eng[e]+=1
  else:
    eng[e]=1
res=sorted(eng.items(),key=lambda x:(-x[1],x[0]))
#-x[1]:將出現頻率取負值，降序排列2.x[0]:當頻率相同，依字母ASCII值進行升序排列
for a,b in res:
  print(a,b)
````
