```python
#c004. 10812 - Beat the Spread!
n=int(input())
for i in range(n):
  a,b=map(int,input().split())
  if a<b or (a+b)%2!=0: #必須偶數
    print("impossible")
  else:
     A=(a+b)//2
     B=(a-b)//2
     print(A, B)
```

```python
#c005. 10300 - Ecological Premium
#每筆獎金算法=((農場的面積/農場裡動物的數目)*農夫的環保等級)*農場裡動物的數目
#整理後就發現動物數目可以消掉，就變成獎金=農場的面積*農夫的環保等級
n=int(input())
for i in range(n):
  ss=0
  f=int(input())
  for j in range(f):
    a,b,c=map(int,input().split())
    s=a*c
    ss+=s
  print(ss)
```

```python
#c014. 10035 - Primary Arithmetic (進位問題)
while True:
  a,b=map(int,input().split())
  if a>b:
    a,b=b,a #由小到大排
  if a==0 and b==0:
    break
  bb=len(str((b)))
  sa=str(a).zfill(bb) #補零
  sb=str(b).zfill(bb)
  s=0
  c=0
  for i in range(1,bb+1):
    if int(sa[-i])+int(sb[-i])+c>=10: #進位
      s+=1
      c=1
    else:
      c=0
  if s==0:
    print("No carry operation.")
  elif s==1:
    print("1 carry operation.")  #單數沒s
  else:
    print(f"{s} carry operations.")
```

```python
#c022. 10783 - Odd Sum
n=int(input())
for i in range(n):
  a=int(input())
  b=int(input())
  s=0  #總和
  for j in range(a,b+1):
    if j%2!=0:
      s+=j
  print(f"Case {i+1}: {s}")
```

```python
#c032. 00382 - Perfection
print("PERFECTION OUTPUT")
while True:
  try:
    num=list(map(int,input().split()))
    for i in num:
      s=0
      if i==0: #先判斷是否結束
        break
      for j in range(1,i):
        if i%j==0:
          s=s+j
      if s==i:
        print(f"{i:5}  PERFECT") #對齊到5格
      elif s<i:
        print(f"{i:5}  DEFICIENT")
      else:
        print(f"{i:5}  ABUNDANT")
  except EOFError:
    break
print("END OF OUTPUT")
```

```python
#c039. 00100 - The 3n + 1 problem
while True: #算出單個數字 n 的完整數列
  try:
    def oup(n):
      opt=[n]
      while True:
        if n==1:
          break
        if n%2==1:
          n=3*n+1
        else:
          n//=2
        opt.append(n)
      return opt
    i,j=map(int,input().split())
    s=0
    for o in range(min(i,j),max(i,j)+1): #排列，避免i,j不是從小到大
      f=len(oup(o))
      if f>s:
        s=f
    print(f"{i} {j} {s}")
  except EOFError:
    break
#目前方法存在可優化處，可使用記憶化遞迴(memoization)解決
#將已計算過的數字長度存在字典中，能避免重複運算
```

```python
#d010. 盈數、虧數和完全數
n=int(input())
s=0
for i in range(1,n):
  if n%i==0:
    s+=i
if s>n:
  print("盈數")
elif s<n:
  print("虧數")
else:
  print("完全數")
```

```python
#d097. 10038 - Jolly Jumpers
while True: #測資多項
  try:
    num=list(map(int,input().split()))
    n=num[0]
    del num[0] #長度
    m=[]
    for i in range(n-1):
      res=abs(num[i]-num[i+1])
      m.append(res)
    if set(m)==set(range(1,n)): #重複的拿掉
      print("Jolly")
    else:
      print("Not jolly")
  except EOFError:
    break
```
