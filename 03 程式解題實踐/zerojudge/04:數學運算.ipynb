````python
#a010. 因數分解
n = int(input())
def factor(n):
  result=[]
  for i in range(2,n+1):
    con=0 #數字出現次數
    while n%i==0:
      n//= i
      con+=1
    if con>0:
      if con==1: #如果次數是1，就只存數字本身
        result.append(str(i))
      else:  #如果次數大於1，存入次方格式
        result.append(f"{i}^{con}")
  return" * ".join(result) #將清單中的所有字串用 * 連接
print(factor(n))
````

````python
#a040. 阿姆斯壯數
n,m=map(int,input().split())
ans=False #如果有→True
for i in range(n,m+1):
  k=len(str(i)) #k位數
  s=0
  for d in str(i): #取每一位數
    s+=(int(d)**k)
  if s==i:
    print(i,end=" ")
    ans=True
if not ans:
  print("none")
````

````python
#a121. 質數又來囉
while True: #注意多個測資
  try:
    a,b=map(int,input().split())
    s=0 #有幾個質數
    for i in range(a,b+1):
      if i<2:
        continue
      if i==2:  #2是質數
        s+=1
      else:
        defi=True
        for j in range(2,int(i**0.5+1)):
          if i%j==0:
            defi=False
            break
        if defi:
          s+=1
    print(s)
  except EOFError:
    break
````

````python
#a158. 11827 - Maximum GCD
import math
N=int(input())
for k in range(N):
  n=list(map(int,input().split()))
  m=0
  for i in range(len(n)): #固定第i個數
    for j in range(i+1,len(n)):
      #從i的下一個數開始，避免了自己和自己比，也避免了重複配對
      ma=math.gcd(n[i],n[j])
      if ma>m: #更新最大值
        m=ma
  print(m)
````

````python
#a410. 解方程 (使用克拉瑪公式)
a,b,c,d,e,f=map(int,input().split())
sx=c*e-b*f
sy=a*f-c*d
s=a*e-b*d
if s==0 and sx!=0 and sy!=0:
  print("No answer")
elif sx==0 and sy==0 and s==0:
  print("Too many")
else:
  x=sx/s
  y=sy/s
  print(f"x={x:.2f}\ny={y:.2f}") #換行、保留兩位小數
````
