```python
#a003. 兩光法師占卜術
m,d=map(int,input().split())
s=(m*2+d)%3
if s==0:
  print('普通')
if s==1:
  print('吉')
if s==2:
  print('大吉')
```

```python
#a004. 文文的求婚
while True:
  try:
    y = int(input())
    output="平年"
    if y%4==0:
      output="閏年"
    if y%100==0:
      output="平年"
    if y%400==0:
      output="閏年"
    print(output)
  except EOFError:
    break
```

```python
#a015. 矩陣的翻轉
while True: #測資有多筆資料
  try:
    r,s=map(int,input().split())
    max=[]
    for i in range(r):
      row=list(map(int,input().split())) #沒有括號 .split()要記得括號
      max.append(row)
    for j in range(s):
      for i in range(r):
        print(max[i][j],end=" ")  #end=" "防止換行
        #先固定j(原本的列)，然後再跑i(原本的行)，
        #就可以讓輸出的順序變成垂直讀取，達成 90 度翻轉的效果。
      print()
  except EOFError:
    break
```

```python
#a104. 排序
while True:
  try:
    n=int(input())
    num=list(map(int,input().split()))
    num.sort() #.sort()預設由小排到大
    print(*num) #星號解包
  except:
    break
```

```python
#c013. 00488 - Triangle Wave
n=int(input())
ny=input() #留一個input()接收測資空白換行
for i in range(n):
  a=int(input())
  f=int(input())
  for j in range(f):
    for k in range(1,a+1): #注意a+1不是a
      print(str(k)*k) #改成字串才能疊加
    for k in range(a-1,0,-1): #倒著回去
      print(str(k)*k)
    print("") #換行
```

```python
#d067. 格瑞哥里的煩惱
y=int(input())
if y%100==0:
  if y%400==0:
    print("a leap year")
  else:
    print("a normal year")
elif y%4==0:
  print("a leap year")
else:
  print("a normal year")
```

```python
#d072(d069) 格瑞哥里的煩惱 (Case 版)
n=int(input())
for i in range(n):
  y=int(input())
  if y%100==0:
    if y%400==0:
      print(f"Case {i+1}: a leap year")
    else:
      print(f"Case {i+1}: a normal year")
  elif y%4==0:
    print(f"Case {i+1}: a leap year")
  else:
    print(f"Case {i+1}: a normal year")
```

```python
#e357. 遞迴函數練習
def F(x):
  if x==1:
    return(1)
  elif x%2==0:
    return F(x/2)
  else:
    return F(x-1)+F(x+1)
x=int(input())
print(F(x))
```
