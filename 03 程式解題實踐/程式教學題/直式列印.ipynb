```python
#直式列印
#改編自2018 資訊學科能力競賽 新北區複賽
n,m=map(int,input().split())
s=[]
for _ in range(n):
  s.append(list(input()))
ss=[["*"]*n for _ in range(m)]
for i in range(m):
  for j in range(n):
    if i < len(s[j]) :  #處理到第 i 列時，如果原本第 j 個字串有足夠的長度，就將該字元填入新矩陣
      ss[i][j] = s[j][i]
for a in ss:
  print(*a,sep="")
```
