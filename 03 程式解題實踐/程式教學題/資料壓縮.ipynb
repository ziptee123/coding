```python
#資料壓縮
#改編自2019 資訊學科能力競賽 彰雲嘉區複賽
abc=list(input())
s=1
for i in range(len(abc)):
  if i < len(abc) - 1 and abc[i] == abc[i+1]:
    s+=1
  elif i == len(abc):
    if abc[i] != abc[i-1]:
      print(f"1{abc[i]}")
  else:
    print(str(s) + abc[i], end="")
    s=1
```
