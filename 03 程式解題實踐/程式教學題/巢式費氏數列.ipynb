```python
#巢式費氏數列
#改編自2019 資訊學科能力競賽 中投區複賽
N=int(input())
def f(x):
  if x==1:
    return 1
  if x==0:
    return 0
  return f(x-1)+f(x-2)
s=[]
for i in range(N):
  m,n=map(int,input().split())
  res=n
  for j in range(m):
    res=f(res)
  s.append(res)
print(*s,sep="\n")
```
