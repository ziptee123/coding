```python
#數字地雷
#改編自2021 資訊學科能力競賽 中投區複賽
n=int(input())
s=[]
for _ in range(n):
  s.append(list(map(int,input().split())))
ss=[[0]*n for _ in range(n)]
d = [(-1, -1), (-1, 0), (-1, 1), (0, -1), (0, 0), (0, 1), (1, -1), (1, 0), (1, 1)]
#周圍位置
for i in range(n):
  for j in range(n):
    for nr,nc in d:
      if 0 <= i + nr < n and 0 <= j + nc < n: #檢查邊界
        if s[i+nr][j+nc] == 1:
          ss[i][j] += 1
for k in ss:
  print(*k) #一行一行輸出
#print(*ss,sep="\n") 這樣輸出是列表形式
```
