```python
#裝置藝術
#改編自2018 資訊學科能力競賽 桃竹苗區複賽
a,b,c=map(int,input().split())
abc=list(map(int,input().split()))
s=0
for i in range(1,len(abc)-1):
  if abc[i]-c<abc[i+1] and abc[i]-c<abc[i-1]:
    continue
  elif abc[i] != max(abc[i],abc[i+1],abc[i-1]):
    continue
  else:
    while True:
      if abc[i]-c<abc[i+1] and abc[i]-c<abc[i-1]:
        break
      abc[i]-=b
      s+=1
print(s)
```
