```python
#自然數的平方根
#改編自2022 資訊學科能力競賽 桃竹苗區複賽
n=int(input())
s=1
i=2
while i*i<=n:
  if n%i**2==0:
    n//=i**2
    s*=i
  else:
    i+=1
if n==1:
  print(s)
elif s==1:
  print(f"sqrt({n})")
else:
  print(f"{s} sqrt({n})")
#我一開始是使用for i in range(2,n**0.5+1)找因數
#但是如果迴圈已經跑過一次因數，不會返回檢查第二次
#因此我改用while搭配if,else，在if那邊不加i+=1，而是等到沒有i**2可以除的時候才加
```
