```python
#移動式鍵盤輸入
#改編自2019 資訊學科能力競賽 中投區複賽
n=int(input())
qaz=[list("QWERTYUIOP"),list("ASDFGHJKL"),list("ZXCVBNM")]
first_r=0
first_c=0
engli=""
def turn(x):
  global first_r,first_c
  if x=="u":
    if first_r>0: first_r-=1
  elif x=="d":
    if first_r<2: first_r+=1
  elif x=="l":
    if first_c>0: first_c-=1
  elif x=="r":
    if first_c<len(qaz[first_r])-1: first_c+=1  #注意len()是數量，從1開始數，所以減1
  if x=="e":
    return qaz[first_r][first_c]
  else:
    return ""  #要回傳空字串，不然engli加none會報錯
for _ in range(n):
  abc=list(input())
  if abc[-3:]!=["e","e","e"]:
    print(-1)
  else:
    for i in range(len(abc)-2):
      engli+=turn(abc[i])
    print(engli)
```
