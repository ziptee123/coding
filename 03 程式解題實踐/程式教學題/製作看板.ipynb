```python
#製作看板
#改編自2021 資訊學科能力競賽 新北區複賽
r,c=map(int,input().split())
eng=list(input())
s=[]
for _ in range(r):
  s.append(list(input()))
dot=sum(k.count(".")for k in s)
space=(dot-len(eng))//2 #前面留幾個點
cou=0
coudot=0
for i in range(r):
  for j in range(c):
    if s[i][j]=="#":
      continue
    else:
      coudot+=1
      if coudot<=space: #先處裡前綴空格
        s[i][j]="."
      elif cou<len(eng):  #只要cou還小於文字長度代表還沒填完，也同時處理了前綴空格問題，不用多寫一個if
        s[i][j]=eng[cou]
        cou+=1
      else:
        s[i][j]="."
for a in s:
  print(*a)
```
