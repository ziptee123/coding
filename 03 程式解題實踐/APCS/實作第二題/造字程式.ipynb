```python
#2023年1月實作題第2題：造字程式
#j606
k,q,r=map(int,input().split())
w=list(input())
s=[]
for i in range(q):
  arra=list(map(int,input().split()))
  ss=["@"]*k
  for j in range(k):
    ss[arra[j]-1]=w[j]
  s.append(ss)
  w=ss
for l in range(r):
  a=""
  for m in range(q):
    a=a+s[m][l]
  print(a)
```
