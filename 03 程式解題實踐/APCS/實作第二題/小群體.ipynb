````python
#2017年3月4日實作題第2題：小群體
#c291
N=int(input())
num=list(map(int,input().split()))
vis=[False]*N
c=0
for i in range(N):
  if not vis[i]:
    c+=1
    cur=i
    while not vis[cur]:
      vis[cur]=True
      cur=num[cur]
print(c)
````

c291 小群體  
這是我的原始程式碼:  
```python
N=int(input())
num=list(map(int,input().split()))
fri=[]
for k in range(N):
  fri.append([k,num[k]])
a=0
sa=0
for i in range(N):
  a,s=fri[i][1],fri[i][1]
  if fri[i][0]==fri[i][1]:
    sa+=1
  for j in fri:
    if j[0]==a:
      a=j[1]
    else:
      continue
    if s==a:
      sa+=1
print(sa)
```
原始邏輯對每一個編號 i 都重新搜尋搜尋一次。  
一個小群體就是一個封閉的環，然而，當程式在i=0的時候計算完了一次小群體，i=1的時候沒有往下算，程式仍然進入相同的環計數。  
這個程式的另一個問題在於for j in fri:只能找到「朋友的朋友」，更長的路徑無法處理。

-------------------------------------------------------------------------------

這題的解決關鍵在於，每個人都指向另一個人，因此整個系統其實一定會形成一個或多個循環。問題本質上就是「數列中有幾個循環」。  
我這題使用的方法是用一個vis列表記錄每個人是否已經被走訪過，避免重複計算同一個循環。  
我從每個還沒走過的人開始，一路追蹤他的好友（cur=num[cur]），直到回到已走過的人為止，這樣就完整走完一個循環。  
而程式每找到一個新的起點就把計數器加1，最後得到總循環數。
