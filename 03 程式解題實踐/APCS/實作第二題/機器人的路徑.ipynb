```python
#2019年6月實作題第2題：機器人的路徑
#e287
n,m=map(int,input().split())
nm=[] #整張地圖
for p in range(n):
  num=list(map(int,input().split()))
  nm.append(num)
minsta=10000000 #初始位置
stx,sty=0,0 #初始座標
for i in range(n):
  for j in range(m):
    if nm[i][j]<minsta:
      minsta=nm[i][j]
      stx=i
      sty=j
vis=[[False]*m for o in range(n)]
vis[stx][sty]=True
total=nm[stx][sty] #所求
x,y=stx,sty
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
while True:
  minnext=10000000
  xnext=-1
  ynext=-1
  for k in range(4):
    nx=x+dx[k]
    ny=y+dy[k]
    if 0<=nx<n and 0<=ny<m:
      if not vis[nx][ny] and nm[nx][ny]<minnext:
        minnext=nm[nx][ny]
        xnext=nx
        ynext=ny
  if xnext==-1: #不用檢查兩個，因為x,y是同時更新的
    break
  x=xnext
  y=ynext
  vis[x][y]=True
  total+=nm[x][y]
print(total)
```
