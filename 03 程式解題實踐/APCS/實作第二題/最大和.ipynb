```python
#2016年10月29日實作題第2題：最大和
#c295
N,M=map(int,input().split())
F=[]
maxi=[]
for k in range(N):
  num=list(map(int,input().split()))
  maxi.append(max(num))
mx=sum(maxi)
for i in maxi:
  if mx%i==0:
    F.append(i)
print(mx)
if len(F)==0:
  print(-1)
  #print(mx\n-1)
else:
  print(*F)
  #print(mx\n*F)
```
