```python
#2024年6月實作題第2題：電子畫布
#o077
#曼哈頓距離=abs(x1-x2)+abs(y1-y2)
h,w,n=map(int,input().split())
canva=[[0]*w for _ in range(h)]
for _ in range(n):
  r,c,t,x=map(int,input().split())
  for i in range(h):  #找每一個點是否小於距離
    for j in range(w):
      dis=abs(i-r)+abs(j-c)
      if dis<=t:
        canva[i][j]+=x
for a in canva:
  print(*a)
```
