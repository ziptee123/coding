```python
#2016年3月5日實作題第2題：矩陣轉換
#b266
r,c,m=map(int,input().split())
s=[]
for _ in range(r):
  ss=list(map(int,input().split()))
  s.append(ss)
inp=list(map(int,input().split()))[::-1] #反轉讀取，因為是從結果推
def turn(s,r,c):
  matrix=[[0 for _ in range(r)] for _ in range(c)]  #將矩陣行列維度互換
  for i in range(c): #注意新矩陣行,列
    for j in range(r):
      matrix[i][j]=s[j][c-1-i] #c-1是行的最大值，-i代表原本增加，新的就會減少，所以減
  return matrix,c,r
def flip(x):
  return x[::-1] #直接翻轉
for a in inp:
  if a==0:
    s,r,c=turn(s,r,c)
  else:
    s=flip(s)
print(r,c)
for b in s:
  print(*b)
```
