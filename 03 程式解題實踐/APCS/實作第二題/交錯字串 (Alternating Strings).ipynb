````python
#2017年10月28日實作題第2題：交錯字串 (Alternating Strings)
#c462
k=int(input())
e=input()
b=[]
c=1
for i in range(1,len(e)): #len(e)剛好就是數字-1
  if e[i].isupper()==e[i-1].isupper():
    c+=1
  else:
    b.append(c)
    c=1
b.append(c)
ans=0
cur=0
for n in b:
  if n==k:
    cur+=k
  elif n>k:
    ans=max(ans,cur+k)
    cur=k
  else:
    cur=0
  ans=max(ans,cur)
print(ans)
````

原本以為可以只數出長度剛好等於k的個數再乘以k即可，   
但是我忽略了當區塊長度大於k時，雖然它不能整塊參與，但它的頭可以跟前面接起來，尾也可以跟後面接起來。  
我也沒有完全中斷我的次數，導致重複累加。

--------------------------------------------------------------------------------
我的改進是透過分類討論(n==k,n>k,n<k)，就能精確處理每一種長度與k的關係。
在每一次迴圈都執行ans=max(ans,cur)，就可以確保捕捉最長的答案
