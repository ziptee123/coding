```python
#2023年6月實作題第2題：特殊位置
#k732
n,m=map(int,input().split())
s=[]
sp=[]
for p in range(n):
  ss=list(map(int,input().split()))
  s.append(ss)
for i in range(n):
  for j in range(m):
    x=s[i][j]
    tsum=0
    for r in range(n):
      for c in range(m):
        if abs(i-r)+abs(j-c)<=x:
          tsum+=s[r][c]
    if tsum%10==x:
      sp.append([i,j])
print(len(sp))
for k in sp:
  print(k[0],k[1])
#這題使用四個for迴圈其實有超時的風險
#可以在第三個迴圈加入if ，將不可能的條件刪除(if r>i+x :break)
```
