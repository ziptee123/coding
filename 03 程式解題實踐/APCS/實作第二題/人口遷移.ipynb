```python
r, c, k, m = map(int, input().split())
s = []
ss = [[0] * c for _ in range(r)]
for _ in range(r):
    s.append(list(map(int, input().split())))
d = [(-1, 0), (1, 0), (0, -1), (0, 1)]
for _ in range(m):
    for i in range(r):
        for j in range(c):
            dif = s[i][j] // k  #要走幾步
            if s[i][j] == -1:   #是-1的就直接不看
                continue
            for nr, nc in d:
                nexr = i + nr
                nexc = j + nc
                if 0 <= nexr < r and 0 <= nexc < c:  #判斷邊界
                    if s[nexr][nexc] == -1:
                        continue
                    else:
                        ss[i][j] -= dif
                        ss[nexr][nexc] += dif
    for a in range(r):
        for b in range(c):
            s[a][b] += ss[a][b]
print(min(x for e in s for x in e if x != -1))
#for e in s:找出每一列
#for x in e:拿出每一個數字
print(max(x for e in s for x in e))
```
