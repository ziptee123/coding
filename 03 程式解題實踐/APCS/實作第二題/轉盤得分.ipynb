```python
#2025年6月實作題第2題：轉盤得分
#q837
m,n,k=map(int,input().split())
s=[]
for _ in range(m):
  ss=list(input())
  s.append(ss)
su=0
for _ in range(k):
  turn=list(map(int,input().split()))
  for j in range(m):
    if turn[j]!=0:
      turn[j]%=n
      s[j]=s[j][-turn[j]:]+s[j][:-turn[j]]
  for i in range(n):
    count={}
    for o in range(m):
      char=s[o][i]
      if char in count:
        count[char]+=1
      else:
        count[char]=1
    maxt=0
    for q in count:
      if count[q]>maxt:
        maxt=count[q]
    su+=maxt
print(su)
```
