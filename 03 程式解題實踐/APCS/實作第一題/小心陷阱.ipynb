```python
#2025年6月實作題第1題：小心陷阱
#q836
p=0
k=int(input())
x1,y1=map(int,input().split())
x2,y2=map(int,input().split())
while True:
  p+=k
  if p%x1==0 and p%x2==0:
    k-=y1+y2
  elif p%x1==0:
    k-=y1
  elif p%x2==0:
    k-=y2
  if k<=0:
    break
print(p)
```
