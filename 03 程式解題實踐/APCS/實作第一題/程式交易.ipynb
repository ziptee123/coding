```python
#2022年1月實作題第1題：程式交易
#h081
n,D=map(int,input().split())
pri=list(map(int,input().split()))
s=0
inb=True
buy=pri[0]###
for i in range(1,n):
  y=pri[i]
  if inb:
    if y>=buy+D:
      s+=y-buy
      buy=y
      inb=False
  else:
    if y<=buy-D:
      buy=y
      inb=True
print(s)
```
