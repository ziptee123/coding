```python
#2020年10月實作題第1題：人力分配
#f312
AO,BO,CO=map(int,input().split())
AT,BT,CT=map(int,input().split())
n=int(input())
max=-float("inf")
for xo in range(n+1):
  xt=n-xo
  yo=AO*xo**2+BO*xo+CO
  yt=AT*xt**2+BT*xt+CT
  ans=yo+yt
  if ans>max:
    max=ans
print(max)
```
