```python
#2019年6月實作題第1題：籃球比賽
#e286
m = 0
o = 0
for i in range(2):
    maj = sum(list(map(int, input().split())))
    out = sum(list(map(int, input().split())))
    print(f"{maj}:{out}")
    if maj > out:
        m += 1
    elif maj < out:
        o += 1
    else:
        m += 1
        o += 1
if m > o:
    print("Win")
elif m < o:
    print("Lose")
else:
    print("Tie")
```
