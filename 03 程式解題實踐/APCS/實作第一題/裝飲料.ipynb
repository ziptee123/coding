```python
#2024年10月實作題第1題：裝飲料
#o711
n=int(input())
wo,wt,ho,ht=map(int,input().split())
wwo=wo**2
wwt=wt**2
vo=wwo*ho
vt=wwt*ht
vs=vo+vt
nv=list(map(int,input().split()))
def tran(v):
  if v>=vs:
    return ho+ht
  else:
    if v>vo:
      return ho+((v-vo)//wwt)
    else:
      return v//wwo
vnow=0
hpre=0
hmax=0
for i in nv:
  vnow+=i
  hnow=tran(vnow)
  hmax=max(hmax,hnow-hpre)
  hpre=hnow
print(hmax)
```
