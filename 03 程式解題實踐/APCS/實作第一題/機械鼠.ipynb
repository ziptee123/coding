```python
#2023年10月實作題第1題：機械鼠
#m370
x,n=map(int,input().split())
food=sorted(map(int,input().split()))
bg=[]
sl=[]
for i in range(n):
  if food[i]>x:
    bg.append(food[i])
  else:
    sl.append(food[i])
if len(bg)>len(sl):
  print(f"{len(bg)} {max(bg)}")
else:
  print(f"{len(sl)} {min(sl)}")
```
