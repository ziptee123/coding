```python
#2021年1月實作題第1題：購買力
#f605(apcs)
n,d=map(int,input().split())
s=0
su=0
for i in range(n):
  a,b,c=sorted(map(int,input().split()))
  if c-a>=d: #原本沒寫=
    s+=1
    su+=(a+b+c)//3
print(f"{s} {su}")
```
