```python
#2022年10月實作題第1題：巴士站牌
#i428
n=int(input())
s=0
m=1000
data=[]
for j in range(n):
  ab=list(map(int,input().split()))
  data.append(ab)
for i in range(n-1):
  sxy=abs(data[i][0]-data[i+1][0])+abs(data[i][1]-data[i+1][1])
  if sxy>s:
    s=sxy
  if sxy<m:
    m=sxy
print(f"{s} {m}")
```
