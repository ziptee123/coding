```python
#2025年1月實作題第1題：等紅綠燈
#q181
g,r=map(int,input().split())
n=int(input())
num=list(map(int,input().split()))
rang=g+r
s=0
for i in num:
  if i%rang>=g:
    s+=r-(i%rang-g)
print(s)
```
