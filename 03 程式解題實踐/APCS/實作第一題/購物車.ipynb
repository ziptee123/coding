```python
#2020年7月實作題第1題：購物車
#f579(apcs)
a,b=map(int,input().split())
n=int(input())
res=0
for j in range(n):
  s=[0]*101
  sh=list(map(int,input().split()))
  for i in sh:
    #if sh[i]==a:
      #s+=1
    #if sh[i]==b:
      #s+=1
    if i>0:
      s[i]+=1
    if i<0:
      s[abs(i)]-=1
  #if s>=2:
    #res+=1
  if s[a]==s[b]:
    if s[a]>0:
      res+=1
  else:
    if s[a]>0 and s[b]>0:
      res+=1
print(res)
```
