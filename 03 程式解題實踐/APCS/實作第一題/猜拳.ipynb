```python
#2020年1月實作題第1題：猜拳
#h026
def rule(a,b):  #F,y
  if (a==0 and b==2)or(a==2 and b==5)or(a==5 and b==0):
    return "A"
  elif a==b:
    return "D"
  else:
    return "B"
F=int(input())
N=int(input())
y=list(map(int,input().split()))
Fhis=[]
winr=0
for i in range(N):
  Fhis.append(F)
  res=rule(F,y[i])
  if res=="A":
    winr+=1
    break
  elif res=="B":
    winr+=1
    break
  else:
    winr+=1
  if i>0 and y[i]==y[i-1]:
    if y[i]==0:
      F=5
    elif y[i]==2:
      F=0
    elif y[i]==5:
      F=2
  else:
    F=y[i]
Fh=" ".join(map(str,Fhis))
if res=="A":
  print(f"{Fh} : Won at round {winr}")
elif res=="B":
  print(f"{Fh} : Lost at round {winr}")
else:
  print(f"{Fh} : Drew at round {winr}")
```
