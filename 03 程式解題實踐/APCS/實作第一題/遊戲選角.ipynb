```python
#2024年1月實作題第1題：遊戲選角
#m931
n=int(input())
ability=[]
for i in range(n):
  a,d=map(int,input().split())
  abil=(a**2)+(d**2)
  ability.append([abil,a,d])
ability.sort(reverse=True)
print(f"{ability[1][1]} {ability[1][2]}")
```
