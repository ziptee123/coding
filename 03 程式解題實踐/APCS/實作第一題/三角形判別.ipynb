```python
#2016年10月29日實作題第1題：三角形辨別
#c294
abc = list(map(int, input().split()))
abc.sort()
print(abc[0], abc[1], abc[2])
if abc[0] + abc[1] <= abc[2]:
    print("No")
elif abc[0]**2 + abc[1]**2 < abc[2]**2:
    print("Obtuse")
elif abc[0]**2 + abc[1]**2 == abc[2]**2:
    print("Right")
else:
    print("Acute")
```
