```python
#2021年11月實作題第1題：修補圍籬
#g595
n=int(input())
h=list(map(int,input().split()))
s=0
for i in range(n):
  if h[i]==0: #原本寫and，這樣後面elif else沒判斷到，變成每一個i都會加一次
    if i==n-1:
      h[i]=h[i-1]
      s+=h[i]
    elif i==0:
      h[i]=h[i+1]
      s+=h[i]
    else:
      s+=min(h[i-1],h[i+1])
print(s)
```
