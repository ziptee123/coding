```python
#2017年10月28日實作題第1題：邏輯運算子 (Logic Operators)
#c461
a,b,r=map(int,input().split())
A=(a!=0)
B=(b!=0)
found=False
if int(A and B)==r:
  print("AND")
  found=True
if int(A or B)==r:
    print("OR")
    found=True
if int(A ^ B)==r:
    print("XOR")
    found=True
if not found:
    print("IMPOSSIBLE")
```
