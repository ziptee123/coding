```python
#2016年3月5日實作題第1題：成績指標
#b964
n=int(input())
res=sorted(map(int,input().split()))
print(*res)
fmax=[s for s in res if s<60]
pmin=[s for s in res if s>=60]
if not fmax:
  print("best case")
else:
  print(max(fmax))
if not pmin:
  print("worst case")
else:
  print(min(pmin))
```
