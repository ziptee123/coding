```python
#2017年3月4日實作題第1題：秘密差
#c290
s=input()
a=0
b=0
for i in range(len(s)):
  if i%2==0:
    a+=int(s[i])
  else:
    b+=int(s[i])
ans=abs(a-b)
print(ans)
```
