```python
#2024年6月實作題第1題：特技表演
#o076
n=int(input())
h=list(map(int,input().split()))
s=[]
for i in range(n-1):
  s.append(h[i]-h[i+1])
lent=0
length=0
for j in range(n-1):
  if s[j]>0:
    length+=1
    if length>lent:
      lent=length
  else:
    length=0
print(lent+1)
```
