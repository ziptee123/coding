```python
#2023年1月實作題第1題：程式考試
#j605
n=int(input())
error=0
max=-1
time=0
for i in range(n):
  a,b=map(int,input().split())
  if b==-1:
    error+=1
  elif b>max:
    time=a
    max=b
total=max-n-(2*error)
if total<=0:
  total=0
print(total,time)
```
