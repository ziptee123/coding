````python
#Q076
a = [3,4,1,5,2,4,2,6,9,7,8]
s = []
for v in a:
  while s and s[-1]>=v:
    s.pop()
  s.append(v)
print(s)
#輸出為何?
#(A)[3,4,5,6,7,8]
#(B)[1,2,4,6,7,8]
#(C)[1,2,6,7,8]
#(D)[1,2,2,6,7,8]
#(E)以上皆非或執行錯誤 Error
#寫題目時忘記while迴圈是持續進行
#如果滿足條件會持續進行，而不是比了一個就停
````

````python
#Q080
def chk(s):
  i,j = 0,len(s)-1
  cost = 0
  while i<j:
    if s[i] != s[j]:
      cost += 1
      s[j] = s[i]
    i,j = i+1,j-1
  return cost
chk('xxarx')
#以下選項中傳入值 s 與回傳值的敘述何者正確?
#(A)chk('xxarx')回傳 2
#(B)chk('abc')回傳 2
#(C)chk('yxarx')回傳 2
#(D)chk('abab')回傳 0
#(E)以上皆不正確
#字串無法更改字元
#如果想更改，要先換成list()
````

````python
#Q095.
def fj(s,d):
  if not s: return ''
  t = ''
  for ch in s[:-1]:
    t += ch+d
  return t+s[-1]
#
s1 = fj('qw','-') #s[:-1]將除了最後一個的部分取出，得q，再加減號，q-後面在加s[-1](w)
s2 = fj(['03','14','2025'],'/') #同上，03/14/2025
s3 = fj(list('post'),'') #取出pos，''是空字串所以t變成pos,加s[-1]變成post
print(s1,s2,s3)
#輸出為何?
#(A)q-w- 03/14/2025/ post
#(B)-qw- /03142025/ ['post']
#(C)q-w 03/14/2025 post
#(D)q-w 03/14/2025 ['post']
#(E)以上皆非或執行錯誤 Error
````

Q101.
```python
a = []
for i in range(n):
  a = [i]+a
```
此程式片段的時間複雜度以 n 表示為下列何者?  
(A)O(n)  
(B)O(nlog(n))   
(C)O(n^2)  
(D)O(n^3)  
(E)以上皆非  
for 算一個n，然而加號算一個複製的動作，故答案為O(n^2)

Q105.
```python
s = 0
for i in range(1,n):
  for j in range(i,n,i):
    s += 1
print(s)
```
此程式片段的時間複雜度以 n 表示,則下列何者為最準確的表示式?   
(A)O(n)   
(B)O(nlog(n))   
(C)O(n2)  
(D)O(n3)  
(E)以上皆非  
for j in range(i,n,i)這串迴圈的執行次數會因為i的增加而減少。   
i=1→跑n次,i=2→跑n/2次
加總所有次數→n+n/2+n/3+...+n/n = n(1/1+1/2+1/3+...+1/n)。  
而數學上，被稱為調和級數，總和會非常接近ln(n)。  
故所求為nlog(n)。

````python
#Q124
s = 'apaltpptpal'
n = len(s)
q = sorted(s) #排序，一樣的字母會在一起
what = cnt = 1
for i in range(1,n):
  if q[i] == q[i-1]: #一樣的就+1
    cnt += 1
  else:
    what = max(what,cnt)
    cnt = 1
print(what) #可知題意為找字母重複最多的數目
#輸出為何?
#(A)1
#(B)2
#(C)3
#(D)4
#(E)以上皆非或執行錯誤 Error
````
