```python
#Q049
a = [2,-3,-1,6,0,3,5,6,8]
t = 0
for v in a:
    if v == 0: break
    if v%2:
      t *= 2
    else:
      v = 0
      t += 1
print(a[0],t)
#輸出為何?
#(A)0 5
#(B)2 5
#(C)2 42
#(D)0 50
#(E)以上皆非或執行錯誤 Error
#後面的else即使改變了v，但也不會影響
#程式執行完v=0後，會繼續往下執行t+=1，回到迴圈開頭，取下一個數字
#所以不會因為else的v=0而直接break結束迴圈
```

```python
#Q053
a, b = [8,4,3,1], [0,4,5]
d = []
for v in b:
  while a and a[-1] <= v:
    d.append(a.pop())
  d.append(v)
d += a
print(d)
#輸出為何?
#(A)[0,1,4,3,5,4,8]
#(B)[0,1,3,4,4,5,8]
#(C)[0,4,5,1,3,4,8]
#(D)[1,3,4,8,0,4,5]
#(E)以上皆非或執行錯誤 Error
#.pop()除了把一個list最後一個元素刪掉之外
#他還可以把那個刪掉的元素傳出來
#所以v=4時，a.pop()會先刪1，然後d.append(1)
```

```python
#Q057
a = [1,2,3,4]
b = a
c = a[:]
b[0] = 0
print(a[0],b[0],c[0])
#輸出為何?
#(A)0 0 1
#(B)1 0 1
#(C)0 0 0
#(D)1 1 1
#(E)以上皆非或執行錯誤 Error
#輸出不是1 1 1的原因是b=a，如果b改變的話a也會跟著變動
#c不動的原因是a[:]只是完整複製了a，但是之後a的改變都沒關係
```

```python
#Q059
total = imax = 0
for i in range(16):
  cnt = 0
  while i>0:
    if i&1: cnt += 1 #i為奇數➔True
    i >>= 1  #等價 i //= 2
  if cnt > imax:
    imax = cnt
  total += cnt
print(imax, total)
#輸出為何?
#(A)5 30
#(B)4 30
#(C)4 32
#(D)4 34
#(E)以上皆非或執行錯誤 Error
```
