```python
#a005
t =int(input()) #幾次
for i in range(t):
  num1,num2,num3,num4=map(int,input().split())
  #先判斷等差，剩下情況都是等比
  if (num2 - num1) == (num3 - num2):
    print(f"{num1} {num2} {num3} {num4} {(num3-num2)+num4}")
  else:
    det=num2//num1 #等比，使用整數除法，不然會有小數點
    print(f"{num1} {num2} {num3} {num4} {det*num4}")
```

```python
#a034
while True:
  try:
    n=int(input())
    if n==0:
      print(0)
    else:
      res=""
      while n>0: #連除
        rem=n%2 #餘數進位
        res=str(rem)+res #用字串形式往後加
        n=n//2
      print(res)
  except EOFError:
    break
```

```python
#a148
while True:
  try:
    num=list(map(int,input().split()))
    n=num[0]
    del(num[0]) #刪除不是分數的
    s=sum(num)/n #計算平均
    if s>59: #及格分>60
      print("no")
    else:
      print("yes")
  except EOFError:
    break
```

```python
#a244
N=int(input())
for i in range(N):
  a,b,c=map(int,input().split())
  if a==1:
    print(b+c)
  elif a==2:
    print(b-c)
  elif a==3:
    print(b*c)
  else:
    print(b//c)
```

```python
#d058(本題不可使用if)
#本題運用了python中布林值參與計算時會變成0,1的特性
n=int(input())
ans=(n>0)-(n<0)
#當n>0時,ans=True-False=1-0=1
#當n=0時,ans=False-False=0-0=0
#當n<0時,ans=False-True=0-1=-1
print(ans)
```
