```python
#羅馬數字轉阿拉伯數字
def roam(s):
    tran = {"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000,"IV":4,"IX":9,"XL":40,"XC":90,"CD":400,"CM":900}
    #字典，羅馬數字和阿拉伯數字的對照表
    #特殊規則:IV、IX這類特殊組合要額外寫出來
    i = 0
    total = 0
    while i < len(s):
        if i + 1 < len(s) and s[i:i+2] in tran:
            #i+1<len(s)是確認下一個字元是否存在，才不會抓到i+1
            #s[i:i+2]則是檢查是否兩個字元的組合有沒有出現在上面的字典裡
            total += tran[s[i:i+2]]
            i += 2
        else:
            total += tran[s[i]]
            i += 1
    return total
#阿拉伯數字轉羅馬數字(使用貪婪演算法)
def roamit(num):
    v = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
    sy = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]
    #將數值由大到小排列，而且順序對應
    #這種設計可以透過同一個索引值i同時存取阿拉伯數字與其對應的羅馬符號
    res = ""
    j = 0
    while num > 0:
    #確保程式會持續處理直到數值被扣完為止
        for k in range(num // v[j]):
            #計算重複次數
            res += sy[j]
            num -= v[j]
        j += 1
    return res

while True:  #主程式碼
    ab = input().strip()
    if ab == "#":
        break
    a, b = ab.split()
    na = roam(a)
    nb = roam(b)
    n = abs(na - nb)
    if n == 0:
        print("ZERO")
    else:
        print(roamit(n))
```
